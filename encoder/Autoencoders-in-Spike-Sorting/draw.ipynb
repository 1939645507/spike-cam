```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.patches as mpatches

# --- 1. 全局样式设置 ---
plt.rcParams['font.family'] = 'serif'
plt.rcParams['font.serif'] = ['Times New Roman'] + plt.rcParams['font.serif']
plt.rcParams['axes.linewidth'] = 1.5

# --- 2. 数据准备 ---
model_names = ['NORA', '$\pi_0$', 'Hyper-VLA (Ours)']
libero_scores = [79.5, 94.2, 97.0]
pro_scores = [39.0, 45.0, 48.0]
colors = ['#EBD9E1', '#C9AFC7', '#A7ACC1']

# --- 3. 开始绘图 ---
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 6), dpi=150)

def draw_styled_bar(ax, labels, scores, title, y_range, y_ticks):
    # 绘制窄柱状图 (width=0.5)
    bars = ax.bar(labels, scores, color=colors, edgecolor='black', linewidth=0.5, width=0.7)
    
    # 设置标题 (大字号)
    ax.set_title(title, fontsize=20, fontweight='bold', pad=25)
    
    # 设置Y轴标签 (大字号)
    ax.set_ylabel('Success Rate (%)', fontsize=18, fontweight='bold')
    
    # 设置坐标轴范围和刻度
    ax.set_ylim(y_range)
    ax.set_yticks(y_ticks)
    
    # 调整刻度数字大小
    ax.tick_params(axis='y', labelsize=20)
    
    # 隐藏上方和右方边框
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    ax.grid(axis='y', linestyle='--', alpha=0.6)
    
    # 设置 X 轴刻度字体 (大字号)
    ax.set_xticklabels(labels, fontsize=16, fontweight='bold')

    # 在柱子顶部标注数值 (增大一点字号)
    for bar in bars:
        height = bar.get_height()
        ax.annotate(f'{height:.1f}%',
                    xy=(bar.get_x() + bar.get_width() / 2, height),
                    xytext=(0, 5), 
                    textcoords="offset points",
                    ha='center', va='bottom', 
                    fontsize=15, fontweight='bold')

# 绘制左图：LIBERO Benchmark
draw_styled_bar(ax1, model_names, libero_scores, 'LIBERO Benchmark', 
                y_range=(60, 105), y_ticks=[60, 70, 80, 90, 100])

# 绘制右图：LIBERO-PRO
draw_styled_bar(ax2, model_names, pro_scores, 'LIBERO-PRO', 
                y_range=(30, 52.5), y_ticks=[30, 35, 40, 45, 50])

# --- 4. 添加统一图例 ---
patches = [mpatches.Patch(color=colors[i], label=model_names[i]) for i in range(3)]
# 图例文字也加大了
fig.legend(handles=patches, loc='upper center', bbox_to_anchor=(0.5, 1.02),
           ncol=3, frameon=True, shadow=True, fontsize=15, prop={'weight':'bold', 'size':15})

plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()
```
